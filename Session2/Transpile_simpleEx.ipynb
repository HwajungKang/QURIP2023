```python
from qiskit import *
```

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.x(1)
qc.cx(0, 1)

qc.draw('mpl')
```

![](cmap.jpeg)

```python
coupling = [[0, 2], [2, 0], [1, 2], [2, 1]]
qc_trans = transpile(qc, coupling_map=coupling, initial_layout=[0, 1])
qc_trans.draw('mpl')
```

```python
qc_trans = transpile(qc, coupling_map=coupling, basis_gates=['x', 'sx', 'id', 'cx', 'rz'])
print(qc_trans.depth())
qc_trans.draw('mpl')
```

```python
qc_trans = transpile(qc, coupling_map=coupling, basis_gates=['x', 'sx', 'id', 'cx', 'rz'], initial_layout=[0, 1])
print(qc_trans.depth())
qc_trans.draw('mpl')
```

```python
qc_trans = transpile(qc, coupling_map=coupling, basis_gates=['x', 'sx', 'id', 'cx', 'rz'], 
                     optimization_level=0)
print(qc_trans.depth())
qc_trans.draw('mpl')
```

```python
qc_trans = transpile(qc, coupling_map=coupling, basis_gates=['x', 'sx', 'id', 'cx', 'rz'], 
                     optimization_level=3)
print(qc_trans.depth())
qc_trans.draw('mpl')
```
